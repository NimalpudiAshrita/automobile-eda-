```python
import pandas as pd
ds= pd.read_csv("automobileEDA.csv")
headers = ds.columns.tolist()
print("automobileEDA.csv", headers)
ds.head()
```

```python
import numpy as np  
ds.replace('?', np.nan, inplace=True)
ds.head()
```

```python
m_s=ds.isnull()
print(m_s.head(5))
```

```python
for colums in m_s.columns.values.tolist():
    print(colums)
    print(m_s[colums].value_counts())
    print("")   
```

Based on the summary above, each column has 205 rows of data, seven columns containing missing data:
normalized-losses: 41 missing data
num-of-doors: 2 missing data
bore: 4 missing data
stroke: 4 missing data
horsepower: 2 missing data
peak-rpm: 2 missing data
price : 4 missing data

```python
#calclating the mean value fo the "normalized losses" column, which contains 41 missing values
#error=('coerce'):Here, if there are any non-numeric values in the "normalized losses" column,
#  they will become NaN, making it easier to handle missing or invalid data.
# inplace=True: This means that the changes will be made directly to the existing "normalized losses" 
# column in the DataFrame ds, rather than creating a new column.
ds["normalized losses"] = pd.to_numeric(ds["normalized losses"], errors='coerce')
avg_norm_loss = ds["normalized losses"].mean()
ds["normalized losses"].replace(np.nan, avg_norm_loss, inplace=True)  
```

```python
# calculating the mean value for the "bore" column, which contains 4 missing values
ds['bore'] = pd.to_numeric(ds['bore'], errors='coerce')
avg_bore = ds['bore'].mean()
ds['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
# calculating the mean value for the "stroke" column, which contains 4 missing values
# here we are using fillna() function instead of replace() function
# fillna() function is used to fill the missing values with a specified value or method.
ds["stroke"] = pd.to_numeric(ds["stroke"], errors='coerce')
avg_stroke = ds["stroke"].mean()
print("Average of stroke:", avg_stroke)
ds["stroke"].fillna(avg_stroke, inplace=True)
```

```python
# calculating the mean value for the "horsepower" column, which contains 2 missing values
# here,we are not using fillna() function we are using replace() function
ds['horsepower'] = pd.to_numeric(ds['horsepower'], errors='coerce')
avg_horsepower = ds['horsepower'].mean()
print("Average horsepower:", avg_horsepower)
ds['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# calculating the mean value for the "peak-rpm" column, which contains 2 missing values
# to numeric() function is used to convert the data type of the column to numeric
ds['peak-rpm'] = pd.to_numeric(ds['peak-rpm'], errors='coerce')
avg_peakrpm = ds['peak-rpm'].mean()
print("Average peak rpm:", avg_peakrpm)
ds['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# replacing the missing values in the "num-of-doors" column with the most frequent value
# because this column is categorical data
# the most frequent value is "four" for which we are replacing the missing values with 'four'
ds['num-of-doors'].value_counts().idxmax()
ds['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
ds.dropna(subset=["price"], axis=0, inplace=True)
ds.reset_index(drop=True, inplace=True)
ds.head()
```

```python
ds.dtypes
```

```python
# here we are converting the data type of the columns to float and integer
# because the data is stored as an object data type
# we are using to_numeric() function to convert the data type of the columns
c_to_f=['bore','stroke','horsepower','peak-rpm','price']
c_to_i=['normalized losses']
for col in c_to_f + c_to_i:
    ds[col]= pd.to_numeric(ds[col], errors='coerce')

# replace nan with  mean of the column
mean_values=ds[col].mean()
ds[col].fillna(mean_values, inplace=True)

# here we are converting the data type of the columns to float and integer after  filling he mean values
ds["normalized losses"]=ds["normalized losses"].astype(int)
print(ds.dtypes)
```

STANDARDIZATION:
Standardiation is the process of transforming data into a common format which allows the recher to make the miningful comparsion.
ex:
Transform mph to L/100km:
L/100km = 235

```python
# here we are converting the data type of the columns to float and integer after  filling he mean values
# we are dividing the "city-mpg" column by 235 to convert it to "city-L/100km"
# 235 is a constant value used for the conversion
ds['city-mpg'] = pd.to_numeric(ds['city-mpg'], errors='coerce')
ds['city-L/100km'] = 235 / ds['city-mpg']
print(ds.head())
```

```python
ds['highway-mpg']= pd.to_numeric(ds['highway-mpg'], errors='coerce')
ds['highway-L/100km']=235/ds['highway-mpg']
ds.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
print(ds.head())
```

NORMALIZATION:

Normalize those variables so their value range from 0 to 1

```python
ds['length']= pd.to_numeric(ds['length'], errors='coerce')
ds['width']= pd.to_numeric(ds['width'], errors='coerce')
 
#NORMALIZATION:
ds['length']= ds['length']/ds['length'].max()
ds['width']= ds['width']/ds['width'].max()

print(ds[['length','width']].head())
```

```python
ds['height']= pd.to_numeric(ds['height'], errors='coerce')
ds['height']= ds['height']/ds['height'].max()
print(ds['height'].head())
```

```python
%pip install scikit-learn

import pickle 
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import  OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
```

```python
ds.info()
```

```python
target = "price"
features = ds.columns.tolist()
features.remove(target)
```

```python
categorical_cols=ds[features].select_dtypes(include=['object']).columns.tolist()
numerical_cols=ds[features].select_dtypes(include=['int64','float64']).columns.tolist()
x=ds[features]
y=ds[target]
# one-hot encoding for categorical columns to change them into numerical columns
preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown='ignore'), categorical_cols)
], remainder='passthrough')

```

```python
#n_estimators: Number of trees in the forest.
#max_depth: Maximum depth of each tree.
#max_features: Number of features considered for splitting at each node.
#random_state: Seed for random number generator for reproducibility.
#min_samples_split: Minimum number of samples required to split an internal node.
#min_samples_leaf: Minimum number of samples required to be at a leaf node.
#n_jobs: Number of jobs to run in parallel. -1 means using all processors.

model = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(
        random_state=42,
        n_estimators=250,
        max_depth=10,
        min_samples_split=4,
        min_samples_leaf=2,
        n_jobs=-1 
    ))
]
)

```

```python
# Train/test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Train
print(" Training RandomForest model...")
model.fit(x_train, y_train)
```

```python

# Remove duplicate columns in x_test if any
x_test = x_test.loc[:, ~x_test.columns.duplicated()]

from sklearn.metrics import mean_absolute_error

y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("\n Model Evaluation Complete")
print(f"R Score: {r2:.4f} ({r2*100:.2f}%)")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
